```python
frame.fillna(value=0).head()
# fillna uzupelnienie brakujacych danych 
# przyjmuje jeden argument ktory zawiera wartosc w kte trzeba wstawic gdzie nie ma wartosci 
# zwraca obiekt dataframe
```

```python
frame['Category'].fillna(value='UNKNOWN',inplace=True)
frame.head()
# 1. [] nazwa kolumnny ktora chcesz wyczyscic 
# 2. fillna na rzecz tej serii wylowaj fillna 
# 3. na liscie argumentow przekazac nowa wartosc ktora ma byc wstawiona wszedzie tam gdzie NaN 
# 4. inplace to jest oryginalny obiekt , bez tego nie zostalo by zadnej zmiany 
```
