```python
pt = sales.pivot_table(index="Retailer country", columns="Year",values="Revenue",aggfunc="sum").head()
pt.head()
# w tej tabeli przestawnej mamy ladnie okreslony index 
```

```python
pt.unstack().head()
# unstacj przelozyc jeden poziom kolumn do poziomu indexu wierszy 
# dane zostaly wyswietlone jako seria danych 
```

```python
pt.unstack().to_frame().head()
# zamienic data series w data frame 
# Rok jest na poczatku i Rtailer country jako druga wartosc w multiindex 
```

```python
pt.unstack().to_frame().swaplevel().head()
# swaplevel na rzec obiektu data frame mozemy przeksztalcic dane + unstack mozey przksztalcic do takiej postaci ktora chilismy  
```

```python
upyt = pt.unstack().to_frame().swaplevel().head()
upyt.columns=["RevenueSum"]
upyt.head()
# Zmiana nazwy 0 na RecenueSum
```

```python
# Kiedy uzywamy melt a kiedy uzywamy unstack 
# Obie polecenie wykonuja podobne dzialalnosci 
### Polecenie melt pracuje z danymi pobranymi z zewnatrz jeszcze nie mam obiektu data frame i kiedy jest multiindex i dane posortowane 
## wczytalem jakies dane ktore maja postac tabeli przedstawnej , kiedy chcemy zmienic do prostej postaci data frame (SWIEZO WCZYTANY PLIK)
### Kiedy pracuje z obiektem data frame ktory juz znajdowal sie w pamieci i dokonalem elne dzialsnosci unstack wtedy uzywamy , unstack jest bardziej wyrozumialy (KIEDY PRACUJE JUZ Z PRACUJACYM PLIKIEM)
```
