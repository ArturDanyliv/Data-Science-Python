```python
cat1.merge(cat2, on = 'CategoryID',suffixes = ['_1','_2'])
# suffixes liste dwoch elementow trzeba okreslic 
# 1 dodawany do kolmnn 1 obiekt dataframe 
# 2 do 2
```

```python
cat1.merge(cat2,on = 'CategoryID',suffixes = ['_1','_2'], how = 'outer')
# inner po czas laczenie danych lewym a prawym wybieramy czesc wspolna 
# how = left zostana wziete wszystkie te wiersze ktore wystepuja w obiekcie po lewej stronie ale nie musza byc znalezione po prawej stronie 
# i to samo z how  = right
# outer czesc wspolna wiersze ktore wystepuja po lewej stronie i po prawej stronie 
```

```python
# wiersz 5 wystepolaw tylko w obiekcie po lewej stronie i 6 
# 5 i 6 wiersz po lewej zostalo znalezione a po prawej nie dla tego jest NaN
# 7 i 8 zostala znaleziona po obu stronach 
# Ostatni wiersz Categorry ID nie wystepowal po lewej stronie byl obecny tylko w cat2 po prawej stronie 
```

```python
cat1.merge(cat2,on = 'CategoryID',suffixes = ['_1','_2'], how = 'outer',indicator = True)
# indicator = True powoduje ze do wynnikowego data frame dodana kolumna merge ktora informuje wzial sie dany wiersz 
# jak bedzie napisane both to znajduje sie po lewej i po prawej stronie 
# left_only 
# right_only 
```

```python
cat_merged = cat1.merge(cat2,on = 'CategoryID',suffixes = ['_1','_2'], how = 'outer',indicator = True)
```

```python
# chce znalezc tylko i wylacznie dane ktore wystepuja po lewej stronie a nie bylo po prawej 
# czy w merge wystepuje left_only 
cat_merged['_merge'] == 'left_only'
```

```python
filter = cat_merged['_merge'] == 'left_only'
cat_merged[filter]
```

```python
filter = cat_merged['_merge'] == 'right_only'
cat_merged[filter]
```

```python
filter = cat_merged['_merge'] == 'both'
cat_merged[filter]
```

```python
cat1.merge(cat2,on = 'CategoryID',suffixes = ['_1','_2'], how = 'left',indicator = True)
# Czy dany wiersz znajdowal sie po lewej stronie jesli tak to bedzie wywietlony nawet jak niew wystepolaw po prawej stronie
# najpierw z lewego szukamy w prawym jak nie to zwraca left_only 
# Category 1 wystepuje tylko po prawej stronie ale nie ma jej po lewej stronie 
# I TAK SAMO Z PRAWYM 
```
