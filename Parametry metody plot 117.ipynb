```python
import pandas as pd 
import numpy as np 
import matplotlib as plt 
%matplotlib inline
```

```python
bar = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/weather_barcelona.csv',index_col='Date')
```

```python
rom = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/weather_rome.csv',index_col='Date')
```

```python
ams = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/weather_amsterdam.csv',index_col='Date')
```

```python
bar.head()
```

```python
temp_max = pd.DataFrame(index = bar.index)
# index taki sam jak obiekt dotyczacy barcelony i ams 
```

```python
temp_max['Barcelona'] = bar['TempMax']
temp_max['Rome'] = rom['TempMax']
temp_max['Amsterdam'] = ams['TempMax']
# max temperatura w poszczegolnych dniach 
```

```python
temp_max.head()
```

```python
temp_max.plot()
# ktory z tych krajow byl najcieplejszy najzimniejszy i jak zmieniala sie temperatura 
# wykres zostal mocno scisniety 
```

```python
temp_max.plot(figsize=(10,5))
# powiekszamy (wieecej miejsca 10 cali szerokosci i 5 cali wysokosci)
```

```python
temp_max.plot(figsize=(15,5), subplots=True)
# zobaczyc osobne wykres ----> subplots = True
```

```python
temp_max.plot(figsize=(15,5), use_index=False)
# use index zawsze prejmuje wartosc True 
# na osi x nalezy umiescic wartosci ktore znajduja sie w index , sens bo pokazywalismy tep zmieniajaca sie w czasie 

# False ----> os x zobaczymy numer kolejny wiersza ktory jest prezentowany na tym wykresie 
# W naszym przypadku analizowalismy 3 miesiace to na osi x wartosci informujace o 90 wierszach dostepnych
# Prezentuje tylko numery wiersszy 
```

```python
temp_max.plot(figsize=(15,5), title = 'Max temperature by city')
# dodana legenda 
```

```python
temp_max.plot(figsize=(15,5), legend=False)
# Zeby wylaczyc legende 
# Na osi y temperatury rozmieszczone sa liniowo w tych samych odstepach -5 0 5 10 15 20
```

```python
temp_max.plot(figsize=(15,5), logy=True)
# zeby wartosci na y byly rozmieszczone logarytmicznie 
# wzloty i upadki sa bardziej scisniety zmienilismy wyswietlenie skali wartosci , wykorzystywamy to kiedy mamy wartosci nie wielkie oraz wartosci bardzo duze 
# nie beda powodowac ze beda pokazane tyko wysokie wartosci a niskie byly by sobie gdzies scisniety w okolicach 0 
# przesledzenie zaleznosci miedzy wartosciami ale z zaburzeniem wartosci odczytywanych wartosci y 
# wlaczenie logarytmicznej osi z  wartosciami

```
