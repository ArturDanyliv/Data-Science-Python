```python
import pandas as pd 
import numpy as np
```

```python
supplies = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/northwind-mongo-master/suppliers _del_1.csv',usecols = ['SupplierID',
'CompanyName','City'])
```

```python
products = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/northwind-mongo-master/products_del_1.csv',usecols = ['ProductID','ProductName','SupplierID'])
```

```python
products.head()
```

```python
prod_and_sup = products.merge(supplies,on="SupplierID")
```

```python
prod_and_sup.head()
```

```python
print(len(supplies))
print(len(products))
print(len(prod_and_sup))
```

```python
prod_and_sup_outer = products.merge(supplies,on="SupplierID",how="outer",indicator=True)
```

```python
prod_and_sup_outer.head()
```

```python
filter_right_only = prod_and_sup_outer["_merge"] == "right_only" 
prod_and_sup_outer[filter_right_only]
```

```python
filter_left_only = prod_and_sup_outer["_merge"] == "left_only" 
prod_and_sup_outer[filter_left_only]
```
