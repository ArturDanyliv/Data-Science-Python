```python
countries = products["Retailer country"].unique()
# countries do ktorej przepisujemy unikalne znalezione kraje
# analiza tylko do takiego jednego fragmentu 
```

```python
countries
```

```python
myOwnGroups = {}
# Przygotujemy zmienna ktora bedzie przechowywala grupy razem z danymi tych grup 
# grupy beda slownikiem 
```

```python
for country in countries:
    myOwnSubDataFrame = products.where(products["Retailer country"] == country).dropna()
    myOwnGroups[country] = myOwnSubDataFrame
# dla kazdego kraju ktory wystepuje na liscie kraje :
#.   tworzymy zupelnie nowy obiekt data frame myOwnSubDataFrame
#       wkladamy wszystkie te produkty ktore spelniaja warunek products retailer country czyli kraj zawarty w takim wierszu ma byc = country pochodzacemu z tabeli countries 
#.      do wyfiltrowania uzywamy metody where 
#.      wynik products[Re co] == country , zwroci serie wartosci True i False 
#.      dropna tylko pokaze True 

# Zapisujemy wynniki 
```
