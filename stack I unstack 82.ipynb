```python
import pandas as pd
import numpy as np 
```

```python
incidents = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/Canadian Railway Crossing Incidents.csv')
incidents.set_index(['Region','EventType'],inplace=True)
incidents.sort_index(inplace=True)
```

```python
incidents.head()
# swaplevel zamienic kolejnosc kolumn 
```

```python
incidents.stack().head()
# stack 
# ---> wziac najbardziej wewnetrzny poziom kolumn i przepisac go do wierszy 
# ---> w naszym przypadku mamy tylko jeden poziom kolumn 
# ---> powstanie indeks ktory sklada sie z 3 segmentow Region EventType i 3 ktory do tej pory jest kolumnami a potem pojawi sie seria danych 
# ---> zawiera te same wartosci ale ina forma 
#  przenieść nagłówki kolumn tak aby stały się kolejnym poziomem w multiindeksie. 
```

```python
incidents.stack().to_frame().head()
# obiekt serii zamieniony na obiekt dataframe z jedna kolumna a indeks 3 kolumnowy 

# z tablicy podlznej robi sie tablica poziomowa 
```

```python
stackedIncidents = incidents.stack().to_frame()
# tworzymy zmienna 
```

```python
stackedIncidents.head()
```

```python
stackedIncidents.unstack().head()
# unstack()
# gdyby zdecedowal sie przewrocic najbardziej wewnetrzny poziom indeksu do kolumn stack 
# najbardziej wewnetrzny poziom indeksu przledrowal do kolumn jak wyzej i nizej 
# Chcesz przenieść najbardziej wewnętrzny poziom multiindeksu tak aby stał się drugim członem w kolumnach.
```

```python
stackedIncidents.unstack().unstack().head()
# multi indeks stal sie normalnym indeksem i pozostala tylko wartosc dla region 
# EventTyp przeledrowala do kolumny 
```

```python
stackedIncidents.unstack(0).head()
# unstack wykonane dla poziomu zerowego 
# na poziomie zerowym byly Region 

# nasze zestawienie alberta British Columbia itd ... prezentuje rodzaje wypadkow na roznego rodzaju przejazdow kolejowych z podzialem na poszczegolne regiony 
```

```python
stackedIncidents.unstack(level = 'Region').head()
# ktory to poziom multi indexu trzeba przelozyc do kolumny mozna uzywac nazwy rowniez 
# 2 sposob piszac level = 
```
