```python
# eplajd urochomic zdefiniowana prze ze mnie funkcje dla kazdego wiersza oddzielnie 
# przyjmuje wiersz row 
# taki wiersz gdzie mozna wydobyc Reporter Type
# wydobyc czesc total operating expenses 

# Definicja funckji 
# tworzymy zmienna reportrtype 
# tworzymy zminna cost ----> i wyrzucamy znak dolar 

# definiujmy zmienna cammand jzeli wiekszy niz 200 000 ---> CLASS A 
# albo vlass b 

# zwracamy wynnik 
def getComment(row):
    reporterTyp = row['Reporter Type']
    cost = float(row['Total Operating Expenses'].replace('$',''))
    
    if(cost >200000):
        comment = 'CLASS A'
    else:
        comment = 'CLASS B'
    return(reporterType+'/'+comment)
```
