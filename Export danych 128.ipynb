```python
import pandas as pd 
import numpy as np 
```

```python
df_blood = pd.DataFrame({'Group'      : ['0', '0', 'A', 'A', 'B', 'B', 'AB', 'AB'],
                        'Rh'          : ['+', '-', '+', '-', '+', '-', '+' , '-'],
                        'Population'  : [31, 6, 32, 6, 15, 2, 7, 1]})
```

```python
df_blood
```

```python
df_blood.to_csv('blood.csv')  # podajemy sciezke dostepu 
# pojawia sie plik blood.csv 

# path nazwa pliku 
# sep = ',' koma zamieniona na cos innego 
# na_rep=''
# columns okreslone kolumny 
# header = True czy musi byc naglowek 
# index = True\False czy musi byc wygenerowany index automatycznie 
# mode = 'w'\'a' w jezeli plik taki juz by istnial to ma zostac nadpisany bez pytan albo a to spowoduje dopisanie do pliku juz ktory istnieje 
# encoding = None polskie znaki zeby mozna bylo odczytac 
# compression = None 
# quotechar=' " ' moga byc przecinki ktore przeszkadazaja z odczytem , cytowanie danych zamkniete w cudzyslowiu 
# decimal = '.' w jaki sposob nalezy pisac liczby dziesietne 
# date_format = None problem z datami albo liczby zmiennoprzenkowe 
```

```python
pd.read_csv('blood.csv')
# pod czas importowania danych index musial byc wygenerowany automatycznie dla tego pojawily sie 2 kolumny index 
```

```python
df_blood.loc[:,"Population"] /= 100
# dzielenie w kolumnie przez 100 wyliczyc procenty 
```
