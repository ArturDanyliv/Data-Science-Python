```python
import pandas as pd
import numpy as np 
import matplotlib as plt 
%matplotlib inline
```

```python
autos = pd.read_csv('/Users/arturdanyliv/Downloads/course-files/autos.csv',encoding='latin-1')
autos.head()
# wystawiane na sprzedaz samochody kodowanie latin-1
# spacje sa zamienione na znak podkresleni w name 
# nazwa cena i rok rejestracji skupiamy sie na tym 
```

```python
# czy uda nam sie znalezc jakies zaleznosci pomiedzy cena a rokiem produkcji samochodu 
# potem mozemy porownac jak poszczegolne marki samochodow utzymuja swoja cene ktore droze a tansze 
# Brak porzadku w kolumnie name tutaj szukamy toyota corolla i Audi a4 
```

```python
import re 
regex = re.compile('.*toyota.*corolla.*', re.IGNORECASE)
filter1  = autos['name'].apply(lambda x: bool(regex.search(x)))
filter2 = autos['price']<12000
filter3 = autos['yearOfRegistration']>1990
tc = autos[filter1 & filter2 & filter3]
# modul re ktory odpowiada za wyrazenia regularne i wszelkie operacje wykonywane z wyrazeniami regularnymi 
# regex wyrazenie regularne ktore bedzie poszukiwane w wierszach (dowolna liczba dowolnych znakow -> .* .*) , toyota , corolla po slowie corolla tez moze byc dowolna ilosc znakow , nie chce przejmowac sie wielkoscia znakow dla tego jest re.IGNORECASE 
# NASTEPNIE 3 SERII wartosci logicznych 
# 1 w tym ogloszeniu jest zeczywiscie sprzedawana toyota corolla (czy nazwy poszczegolnych aukcij spelniaja warunek wyrazenia regularnego )
# 2 mniejsze 12000
# 3 starsze niz 1990
```

```python
import re 
regex = re.compile('.*Audi. *a4.*', re.IGNORECASE)
filter1 = autos['name'].apply(lambda x: bool(regex.search(x)))
filter2 = autos['price']<12000
filter3 = autos['yearOfRegistration']>=1990
audi = autos[filter1 & filter2 & filter3] 
```

```python
tc.plot.hexbin(x='yearOfRegistration',y='price',gridsize=10,title='Toyota')
audi.plot.hexbin(x='yearOfRegistration',y='price',gridsize=10,title='Audi')
# hexbin wykres szesciokaty 
# gridsize jak najmniejsze tym wieksze szesciokaty 
```

```python
tc_counts = tc['yearOfRegistration'].value_counts().sort_index()
```

```python
tc_counts.head()
```

```python
audi_counts = audi['yearOfRegistration'].value_counts().sort_index()
```

```python
audi_counts.head()
```

```python
car_counts = pd.DataFrame(index = tc_counts.index.append(audi_counts.index).unique())
# ilosc samochodow toyota i ilosc samochodow Audi nasumowana 
# suma indexow dla samochodow toyota i Audi 
```

```python
car_counts['Toyota'] = tc_counts
# kopiujemy wartosci 
```

```python
car_counts['Audi'] = audi_counts
# kopiujemy wartosci 
```

```python
car_counts.head()
```

```python
car_counts.fillna(0,inplace=True)
# pozbuwamy sie wartosci NA
```

```python
car_counts.sort_index(inplace=True)
# sortujemy dane w kolejnosci inexu czylo roku produkcji samochodu 
```

```python
car_counts.plot()
# pierwszy wykres ktory prezentuje ilosc sprzedawanych samochodow 
```

```python
car_counts.plot(kind='area')
# zmieniamy wykres 
# wykres ma byc typu area
```
