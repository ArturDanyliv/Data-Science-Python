```python
# # sprobujmy sortowanie po EventType 
# ten index rowniez nie jest unikalny 
# jedna kolumna nie wystrczy tu do poindeksowanie tego zbioru danych 
incidents.set_index('EventType').head()
```

```python
incidents.set_index(['Region','EventType']).head()
# keys  tych kluczy budjace index moze byc wiecej ---> korzystamy z listy 
# przyjelo liste nazw dwoch kluczy dzila 
# set_index ma domyslnie ustawione inplace=False 
```

```python
# zmieniamy kolejnosc budowania indexu 
# rowniez inplace ma byc=True 

# nic sie nie wyswietli bo mamy wpisac incidnts.head()
incidents.set_index(keys=['EventType','Region'],inplace=True)
```

```python
incidents.head()
# dwie kolummny sa pogrubiona scianka 

# sposob wyswietlenie zmienilo sie wczesniej widzielismy kazdy z regionow i nazwa wyswietlona tylko jeden raz 
#.  ten region mogl zawierac kilka pozycji opisujace roznego rodzaju zdarzenia EventType i wtedy byly wypisywane jedno pod drugim tak dlugo az jak przeslismy do nowego regionu 
#   i w tym nowym regionu zniwyu powtarzaly sie wen Type

# dane trzeba posortowac wedlug indexu 
```

```python
# dane trzeba posortowac wedlug indexu 
incidents.sort_index(inplace=True)
```

```python
incidents.head(10)
# tym pierwszym ktory buduje indx jest EventType
# w pierszej kolejnosci Acc i regiony 

# Sortowanie jest w kolejnosci rosnacej A B C 
```

```python
# Zeby EvntType byl posortowany w kolejnosci rosnacej 
# Regiony pokazywaly siie w kolejnosci malejacej 
# accnding=True 

# przkacujemy liste do ascending ktora sklada sie z wartosci logicznych 

# Pierwsza wartosc True okresla ze pierwszy klucz w indeksie ma byc posortowany rosnaco 
# druga wartosc False okresla ze drugi klucz indeksu czyli region maja byc posortowane w kolejnosci malejacej 
incidents.sort_index(ascending=[True,False],inplace=True)
```
